# 数据介绍

```python
'''
S satisfaction 满意度 0~1
LPE  last project evaluation by a client 客户对最近一次项目的评价0~1
NP number of projects 去年从事的项目数量
ANH average number of hours worked in the last 12 month by the employee 去年的平均工作时长
TIC time spent in the company in years by the employee 去年在公司的时长
Newborn 去年是否生孩子， 0，1
'''
hr_data.head()
```

```python
hr_data.describe()
```

# 数据处理

```python
# 归一化 z = (x - u) / s
from sklearn import preprocessing 
test_data = hr_data.copy()
hr_scale = preprocessing.scale(test_data)
```

# 数据建模

# **掌握轮廓系数的计算**

```python
# 确定分类个数 dendrogram
import plotly.figure_factory as ff
fig = ff.create_dendrogram(hr_scale)
fig.update_layout(width=1000, 
                 height=1000)
fig.add_trace(go.Scatter(y=[4.6]*2000, x=[x for x in range(1, 2000)]))
fig.show()
```

```python
# elbow method
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 30):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(hr_scale)
    wcss.append(kmeans.inertia_)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=[x for x in range(1, 30)], y=wcss, marker=dict(size=300)))
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot()
axes.plot(range(1,30), wcss, marker="o")
```

```python
# kmeans建模
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
cluster = kmeans.fit(hr_scale)
cls = cluster.predict(hr_scale)
cls
```

```python
set(cls)
```

```python
for i in np.unique(cls):
    print('cls={} : {}'.format(i, np.sum(cls==i))  )
```

```python
# hc建模
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=4 , affinity='euclidean', linkage='ward')
hc_clus = hc.fit_predict(hr_scale)
```

```python
hc_clus
```

```python
for i in np.unique(hc_clus):
    print('cls={} : {}'.format(i, np.sum(hc_clus==i))  )
```
