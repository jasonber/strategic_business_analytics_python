```python
# 划分数据集
X = hr_df.iloc[:,:-1].values
y = hr_df.iloc[:,-1].values
```

```python
# statsmodels 建模
X2 = hr_df.iloc[:,:-1].values
y2 = hr_df.iloc[:,-1].values
from statsmodels.api import Logit # Logit model https://baike.baidu.com/item/Logit%E6%A8%A1%E5%9E%8B/7286579?fr=aladdin
lr2 = Logit(y2, X2)
res2 = lr2.fit()
```

```python
res2.summary()
```

```python
# 数据sacle
from sklearn import preprocessing
scale = preprocessing.StandardScaler()
X[:,:-1] = scale.fit_transform(X[:,:-1])

lr = Logit(y, X)
res = lr.fit()
```

```python
res.summary()
```
